```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from bertviz import head_view

model_path = "./final_emotion_model" 
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path, output_attentions=True)
model.eval()

sentence = "I am really excited with how things turned out"

inputs = tokenizer.encode_plus(sentence, return_tensors="pt")
input_ids = inputs["input_ids"]

outputs = model(**inputs)
attention = outputs.attentions

head_view(attention, tokenizer.convert_ids_to_tokens(input_ids[0]))
```
